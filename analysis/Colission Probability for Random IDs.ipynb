# Collision Probability for Randomly Selected Labels

If $f$ clients independently choose labels from $N$ choices, the probabiltiy that at least two chosen labels are identical is
$$ P[\text{collision}] = 1 - P[\text{all unique}] = 1 - \frac{N}{N} \cdot \frac{N-1}{N} \cdot \frac{N-2}{N} \cdots \frac{N-f+1}{N} = 1 - \prod_{j=0}^{f-1}\frac{N-j}{N}. $$
This notebook visualizes this collision probabilty for some values of $f$ and $N$.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fs = np.flip(np.arange(0, 20, 1))
Ns = 2**np.arange(3, 12)
```

```python
Pcoll = np.empty(shape=(len(fs), len(Ns)))

for fi, f in enumerate(fs):
    for Ni, N in enumerate(Ns):
        Pcoll[fi, Ni] = 1 - np.prod([(N-j)/N for j in range(0, f)])
```

```python
plt.figure(figsize=(14, 6))
sns.heatmap(Pcoll, xticklabels=Ns, yticklabels=fs, cmap='icefire', annot=True)
plt.xlabel('Available IDs')
plt.ylabel('Client Query Frequency')
plt.title('Collision Probability')
```
